```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
# origin is in the center, columns are x coordinate, rows are y coordinate
# x increases towards the right, y increases upwards
x_sz = 200
y_sz = 200

# one voxel that should be interested about stimuli in the upper left quadrant
x = -30
y = 50
sigma = 40

mu = np.array([y, x])

sigma = np.eye(2) * sigma**2
```

```python
ys, xs = np.mgrid[y_sz:0:-1, 0:x_sz]
ys -= y_sz // 2
xs -= x_sz // 2

x1d = xs.ravel()
y1d = ys.ravel()
```

```python
plt.figure(); plt.matshow(ys)
plt.figure(); plt.matshow(xs)
```

```python
# define gaussian distribution object
mvn = stats.multivariate_normal(mu, sigma)
```

```python
# evaluate probability distribution function of this object at all positions in 
# pixel space
coord_pairs = np.vstack((y1d, x1d)).T
vals1d = mvn.pdf(coord_pairs)

# need to reshape to matrix
vals = vals1d.reshape(x_sz, y_sz)
```

```python
# enjoy
plt.matshow(vals)
```

# Compute pRFs for multiple voxels

```python
n_voxels = 10000

mus = np.array([
    np.random.randint(ys.min(), ys.max()+1, n_voxels),
    np.random.randint(xs.min(), xs.max()+1, n_voxels)])

sigmas = np.random.randint(1, np.min([x_sz, y_sz]), (1, 1, n_voxels)) ** 2

# first voxel should be same as in the example above
mus[:, 0] = [y, x]
sigmas[0, 0, 0] = sigma[0, 0]

# invert all covariance matrices in one go using the fact that the inverse of a
# diagonal matrix is a diagonal matrix with diagonal elements replaced by the
# reciprocals of the original diagonal matrix
sigmas_inv = np.stack([np.eye(2)] * n_voxels, 2) / sigmas
```

```python
# turn all points in the image into an array of x, y coordinate pairs and
# and expand with the number of voxels
X = np.stack([coord_pairs] * n_voxels, 2) 

# subtract the pRF means from each of these coordinate pairs so that
# coordinates are relative to pRF centers
delta_X = X - mus
```

```python
# calculate first part of exponent in multivariate PDF for all voxels at all
# image locations, i.e.:
#
# (x - mu)T * S^{-1} ...
#

F1 = np.einsum('ij...,jk...', delta_X, sigmas_inv)
```

```python
# proceed with the second part of the calculation in the exponent, i.e.:
#
# ... F1 * (x - mu)
#

F2 = np.einsum('...i,...i', F1, delta_X.transpose(2, 0, 1))
```

```python
# divide exponent by -2 and compute power
pdfs = np.exp(-.5 * F2)
```

```python
# instead of calculating the factor outside the power, simply rescale PDFs so
# that they integrate to 1.
pdfs = pdfs / np.tile(pdfs.sum(axis=1), (x_sz * y_sz, 1)).T
```

```python
# verify that the first voxel still has the intended pRF function
plt.matshow(pdfs[0].reshape(y_sz, x_sz))
```

```python
pdfs.sum(axis=1)
```

```python
# check if PDF of first voxel is indeed the same by plotting values against 
# each other and verifying that they fall on the orange identity line
plt.scatter(pdfs[0], vals.ravel() / vals.sum())
plt.plot([0, pdfs[0].max()], [0, pdfs[0].max()], color='orange')
```
